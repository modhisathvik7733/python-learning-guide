# 10 - Seaborn Objects API (The "Next-Gen" Interface)

## Goal
Introduction to `seaborn.objects` (imported as `so`), a new, more consistent API inspired by the Grammar of Graphics (like ggplot2).
- **Interface**: `so.Plot(data, x, y)...`
- **Method Chaining**: `.add(mark, stat, orient)...`
- **Composability**: Easily layering different marks (dots, lines, areas).
- **Stat Transforms**: Aggregating or fitting models directly in the plot spec.

## Why use it?
- More flexible than the classic functional API (`sns.scatterplot`, `sns.barplot`).
- Better separation of data mapping, statistical transformation, and visual representation.
- Fully object-oriented and type-hinted.

## Quick concept map
| Concept | Classic API | Objects API |
|---|---|---|
| Initialization | `sns.scatterplot(data=df, ...)` | `p = so.Plot(df, ...)` |
| Adding layers | Overplotting with multiple calls | `.add(so.Dot()).add(so.Line())` |
| Faceting | `FacetGrid(...)` | `.facet(col=...)` |
| Scaling | arguments like `log_scale=True` | `.scale(x='log')` |
| Themes | `sns.set_theme()` | `.theme(...)` |

## Setup


```python
import seaborn.objects as so
import seaborn as sns
import pandas as pd
import numpy as np

# Load standard datasets
penguins = sns.load_dataset('penguins').dropna()
tips = sns.load_dataset('tips')
```

## 1) Basic Syntax: The `Plot().add()` pattern
1. `so.Plot(data, x, y)`: Define data and coordinates.
2. `.add(mark)`: Define the geometric shape (Dot, Bar, Line).


```python
# Basic scatter plot
(
    so.Plot(penguins, x="bill_length_mm", y="bill_depth_mm")
    .add(so.Dot())
)
```

## 2) Mapping Properties (Color, Size, etc.)
Map data variables to visual properties inside `Plot()` or `.add()`.


```python
# Scatter with color and point size mapped to columns
(
    so.Plot(penguins, x="bill_length_mm", y="bill_depth_mm", color="species", pointsize="body_mass_g")
    .add(so.Dot(alpha=0.7))
)
```

## 3) Statistical Transforms (`so.Agg`, `so.PolyFit`)
Perform aggregations or model fitting on the fly.


```python
# Bar plot with aggregation (mean by default) and Error Bars
(
    so.Plot(tips, x="day", y="total_bill")
    .add(so.Bar(), so.Agg())            # Mean bar
    .add(so.Range(), so.Est(errorbar="sd")) # Standard deviation range
)
```

```python
# Scatter + Linear Fit (PolyFit order=1)
(
    so.Plot(tips, x="total_bill", y="tip")
    .add(so.Dot(alpha=0.5))
    .add(so.Line(color="red"), so.PolyFit(order=1))
)
```

## 4) Faceting (`.facet()`)
Create small multiples easily.


```python
(
    so.Plot(tips, x="total_bill", y="tip", color="smoker")
    .add(so.Dot())
    .facet(col="time", row="sex")
)
```

## 5) Customization (Scales & Theme)


```python
(
    so.Plot(penguins, x="body_mass_g", y="flipper_length_mm", color="species")
    .add(so.Dot(pointsize=4))
    .scale(x="log", color="viridis") # Log scale x, viridis color map
    .label(x="Body Mass (g) - Log Scale", title="Penguin Size Analysis")
    .theme({"axes.grid": True})
)
```
