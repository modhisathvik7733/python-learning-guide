```python
## Two pointer Convergence Pattern
```

```python
def two_pointer_convergence(arr,target):
    left = 0
    right = len(arr)-1
    while left < right :
        current = arr[left] + arr[right]
        if current == target:
            return [left,right]
        elif current < target:
            left+=1
        else:
            right-=1
    return None
```

```python
arr = [1,2,3,4,5]
two_pointer_convergence(arr,8)
```
