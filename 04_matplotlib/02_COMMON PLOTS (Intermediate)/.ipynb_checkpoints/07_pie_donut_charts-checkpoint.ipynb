# Pie & Donut Charts

## Overview

**Pie charts** show parts of a whole as slices of a circle.

### What We'll Learn

**1. Basic Pie Charts** 🥧
- Creating pie charts
- Labels and percentages
- Colors and styling
- Exploding slices

**2. Donut Charts** 🍩
- Creating donut charts
- Center text
- Multiple rings
- Advanced styling

**3. Advanced Techniques** 🚀
- Nested pie charts
- Auto-percentages
- Custom wedge properties
- Shadow and 3D effects

### Why Use Pie Charts?

```
✓ Show part-to-whole relationships
✓ Easy to understand
✓ Good for few categories (2-6)
✓ Visual impact
✗ Bad for many categories
✗ Hard to compare similar values
✗ Consider bar charts instead
```

### Common Use Cases

- Market share
- Budget allocation
- Survey results
- Demographic breakdowns
- Resource distribution

## 1. Basic Pie Charts

### Basic Syntax

```python
# Simple pie
ax.pie(sizes, labels=labels)

# Full parameters
ax.pie(sizes,
      labels=labels,           # Slice labels
      autopct='%1.1f%%',      # Percentage format
      startangle=90,          # Rotation angle
      explode=(0, 0.1, 0),    # Explode slices
      colors=colors,          # Custom colors
      shadow=True,            # Add shadow
      wedgeprops={'edgecolor': 'black'},  # Edge style
      textprops={'fontsize': 12})  # Text style
```

### Donut Chart (Pie with Center Hole)

```python
# Create pie chart
wedges, texts, autotexts = ax.pie(sizes, labels=labels, 
                                   autopct='%1.1f%%')

# Add center circle (makes it donut)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
ax.add_artist(centre_circle)

# Or use wedgeprops
ax.pie(sizes, wedgeprops={'width': 0.3})  # Donut width
```

### Exploding Slices

```python
# Explode specific slices
explode = (0, 0.1, 0, 0)  # Explode 2nd slice
ax.pie(sizes, explode=explode, labels=labels)
```

### Auto-Percentage Formatting

```python
autopct='%1.1f%%'     # One decimal: 15.5%
autopct='%1.0f%%'     # No decimal: 16%
autopct='%.2f%%'      # Two decimals: 15.50%

# Custom function
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val:d})'
    return my_autopct

ax.pie(sizes, autopct=make_autopct(sizes))
```

### Best Practices

```
✓ Limit to 5-6 slices maximum
✓ Sort slices by size
✓ Use clear labels
✓ Add percentages
✓ Use distinct colors
✓ Consider donut for modern look
✗ Don't use for many categories
✗ Don't use 3D (distorts perception)
✗ Don't compare multiple pies
```

````python
print("=== PIE & DONUT CHARTS ===\n")

# Example 1: Basic pie chart
print("Example 1: Market Share Pie Chart")

fig, ax = plt.subplots(figsize=(10, 8))

sizes = [35, 25, 20, 15, 5]
labels = ['Company A', 'Company B', 'Company C', 'Company D', 'Others']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
explode = (0.1, 0, 0, 0, 0)  # Explode 1st slice

wedges, texts, autotexts = ax.pie(sizes, 
                                   explode=explode,
                                   labels=labels,
                                   colors=colors,
                                   autopct='%1.1f%%',
                                   startangle=90,
                                   wedgeprops={'edgecolor': 'white', 'linewidth': 2})

# Style percentages
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

ax.set_title('Market Share Distribution', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Example 2: Donut chart
print("\n" + "="*70)
print("Example 2: Budget Allocation Donut Chart")
print("="*70)

fig, ax = plt.subplots(figsize=(10, 8))

sizes = [40, 30, 20, 10]
labels = ['Marketing', 'R&D', 'Operations', 'Admin']
colors = ['#E64B35', '#4DBBD5', '#00A087', '#3C5488']

wedges, texts, autotexts = ax.pie(sizes,
                                   labels=labels,
                                   colors=colors,
                                   autopct='%1.1f%%',
                                   startangle=90,
                                   wedgeprops={'width': 0.4, 'edgecolor': 'white', 'linewidth': 2})

# Style text
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

# Center text
ax.text(0, 0, 'Budget\n$100M', ha='center', va='center', 
       fontsize=16, fontweight='bold')

ax.set_title('Department Budget Allocation', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n💡 Tip: Donut charts have modern, clean aesthetic")
````
