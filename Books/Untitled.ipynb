```python
import numpy as np
import matplotlib.pyplot as plt

# Your data
data = np.array([15, 20, 35, 40, 50, 25, 30, 45, 50, 55])

# Calculate histogram with bins
values, base = np.histogram(data, bins=10)
print(values)
print(base)
# Calculate cumulative frequency
cumulative = np.cumsum(values)
print(cumulative)
# Plot the ogive
plt.plot(base[:-1], cumulative, 'ro-')  # red circles with line
plt.xlabel('Data Values')
plt.ylabel('Cumulative Frequency')
plt.title('Ogive Graph')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Math', 'Physics', 'Chemistry']
student_scores = [85, 78, 92]
max_scores = [95, 88, 98]

x = np.arange(len(categories))  # Label positions
width = 0.35  # Bar width

fig, ax = plt.subplots()
bar1 = ax.bar(x - width/2, student_scores, width, label='Student')
bar2 = ax.bar(x + width/2, max_scores, width, label='Max Score')

ax.set_xlabel('Subjects')
ax.set_ylabel('Scores')
ax.set_title('Student vs Max Score Comparison')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

```python
import pandas as pd

# Step 1: Create the contingency table from your image [file:47]
data = [
    ['', 'Football', 'Not Football', 'Total'],
    ['Male', 26, 24, 50],
    ['Female', 12, 38, 50],
    ['Total', 38, 62, 100]
]

df = pd.DataFrame(data[1:], columns=data[0])
print("Contingency Table:")
print(df.to_markdown(index=False))
print("\n" + "="*50 + "\n")

# Step 2: Extract counts for probability calculation
total_people = 100
count_male = 50
count_football = 38
count_male_football = 26  # Intersection (Male AND Football)

# Step 3: Calculate individual probabilities
p_male = count_male / total_people
p_football = count_football / total_people
p_both = count_male_football / total_people

# Step 4: Union formula: P(Male OR Football) = P(Male) + P(Football) - P(Male AND Football)
p_male_or_football = p_male + p_football - p_both

# Alternative pandas method (using your table data)
df_table = pd.DataFrame({
    'Gender': ['Male']*50 + ['Female']*50,
    'Football': [1]*26 + [0]*24 + [1]*12 + [0]*38
})
p_male_or_football_pandas = len(df_table[(df_table['Gender'] == 'Male') | (df_table['Football'] == 1)]) / len(df_table)

print(f"P(Male) = {p_male *100:.2f}")
print(f"P(Football) = {p_football:.2f}")
print(f"P(Male AND Football) = {p_both:.2f}")
print(f"P(Male OR Football) = {p_male_or_football:.2f}")
print(f"Pandas verification: {p_male_or_football_pandas:.2f}")
```
